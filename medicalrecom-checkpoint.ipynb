```python
#test train

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
encoder.fit(y)
y = encoder.transform(y)
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=42)
```

```python
print("X_train Shape:",X_train.shape)
print("X_test Shape:",X_test.shape)
print("y_train Shape:",y_train.shape)
print("y_test Shape:",y_test.shape)
```

```python
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score, confusion_matrix

models = {
    "SVC":SVC(kernel='linear'),
    "Random Forest Classifier":RandomForestClassifier(n_estimators=100,random_state=100),
    "Decision Tree Classifier":DecisionTreeClassifier(),
    "Gradient Boosting Classifier":GradientBoostingClassifier(n_estimators=100,random_state=100),
    "Multinomial NB":MultinomialNB(),
    "K-Neighbors Classifier":KNeighborsClassifier(n_neighbors=5)
}

for model_name, model in models.items():
    model.fit(X_train,y_train)
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test,y_pred)
    mat = confusion_matrix(y_test,y_pred)
    
    # Calculate confusion matrix
    print(f"{model_name} Accuracy: {acc}")
    print(f"{model_name} Confusion Matrix:")
    print(np.array2string(mat, separator=', '))
    print("=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=")
```

```python
import pickle
import pandas as pd

# Load trained model
model = pickle.load(open("model/svc.pkl","rb"))
description_df = pd.read_csv(r"C:\Users\potht\Medi_project\Datasets\description.csv")
diets_df = pd.read_csv(r"C:\Users\potht\Medi_project\Datasets\diets.csv")
medications_df = pd.read_csv(r"C:\Users\potht\Medi_project\Datasets\medications.csv")
precautions_df = pd.read_csv(r"C:\Users\potht\Medi_project\Datasets\precautions_df.csv")
workout_df = pd.read_csv(r"C:\Users\potht\Medi_project\Datasets\workout_df.csv")
```

```python
def predict_disease(sample_symptoms):
    
    # Predict disease
    prediction = model.predict([sample_symptoms])
    disease = encoder.inverse_transform(prediction)[0]

    print("Predicted Disease:", disease)

    # Description
    description = description_df[description_df['Disease']==disease]['Description'].values
    if len(description) > 0:
        print("\nDescription:", description[0])

    # Precautions
    precautions = precautions_df[precautions_df['Disease']==disease].iloc[:,1:].values
    if len(precautions) > 0:
        print("\nPrecautions:")
        for i in precautions[0]:
            print("-", i)

    # Medications
    medications = medications_df[medications_df['Disease']==disease]['Medication'].values
    if len(medications) > 0:
        print("\nMedications:")
        for i in medications:
            print("-", i)

    # Diet
    diet = diets_df[diets_df['Disease']==disease]['Diet'].values
    if len(diet) > 0:
        print("\nRecommended Diet:")
        for i in diet:
            print("-", i)

    # Workout
    workout = workout_df[workout_df['disease'] == disease_name]['workout'].values
    if len(workout) > 0:
        print("\nWorkout:")
        for i in workout:
            print("-", i)
```

```python
sample = X_test.iloc[0].values

predict_disease(sample)
```
